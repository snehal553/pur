### Scaling 

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
x = df.drop('Purchased', axis = 1)
y = df['Purchased']
```

```python
x_train, x_test, y_train,y_test = train_test_split(x,y, test_size=0.25, random_state=5, stratify=y)
```

```python
std_scale = StandardScaler()
x_train_scale =std_scale.fit_transform(x_train) 
```

```python
x_train_scale
```

```python
x_test_scale = std_scale.transform(x_test)
```

## Model Training

```python
from sklearn.svm import SVC
```

```python
model = SVC()
model.fit(x_train_scale, y_train)
```

```python
### Model EValuation 
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
y_pred = model.predict(x_test_scale)
y_pred

def model_eval(actual,pred):
    acc = accuracy_score(actual,pred)
    print(f"Accuracy Score = {acc}")
    
    cm = confusion_matrix(actual,pred)
    print(cm)
    
    cr = classification_report(actual, pred)
    print(cr)
    
    return "Success"
model_eval(y_test, y_pred)
```

```python
### Training data evaluation 

y_pred = model.predict(x_train_scale)
model_eval(y_train,y_pred)
```

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
import numpy as np
```

```python

svc = SVC()
hyp = {
    "C" : np.arange(1,11),
    "kernel" : ['linear', 'poly','rbf','sigmoid']
}

gscv = GridSearchCV(svc, hyp, cv = 5)
gscv.fit(x_train_scale, y_train)
```

```python
gscv.best_params_
```

```python
op_model = SVC(C= 6, kernel='rbf')
op_model.fit(x_train_scale,y_train)

### Model Evalaution 
### Testing 
y_pred = op_model.predict(x_test_scale)
model_eval(y_test,y_pred)

### Training 
y_pred = op_model.predict(x_train_scale)
model_eval(y_train,y_pred)
```

```python
### User Defined 

gender = 'Male'
age = 35
salary = 50000

def predict(gender, age, salary):
    '''Predict function op_model '''
    
    # converting categorical data into numerical data 
    if gender == 'male'  or gender == "Male" or gender == "MALE" or gender == "M":
        gender = 1
    else: 
        gender = 0 
    
    ## Creating an array and scaling
    arr = np.array ([gender, age, salary])
    arr_scale = std_scale.transform([arr])
    predict = op_model.predict(arr_scale)
    predict

    if predict[0] == 0: 
        print(f"Not Purchsed the product")
    else: 
        print(f"Purchase the  Product")
        
    return predict[0]

predict(gender,age,salary)
```
